```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times']
# plt.rcParams['font.sans-serif'] = 'Times-Roman'

p256 = [0 for x in range(0, 100)]
x25519 = [0 for x in range(0, 100)]
eng25519 = [0 for x in range(0, 100)]
all_openssl = [0 for x in range(0, 100)]

u_p256 = [0 for x in range(0, 100)]
u_x25519 = [0 for x in range(0, 100)]
u_eng25519 = [0 for x in range(0, 100)]
u_all_openssl = [0 for x in range(0, 100)]

i = 0
with open('handshake/server_p256.txt', 'r') as f1:
    for line in f1:
        p256[i] = 10000000000 / int(line[42:49])
        i = i + 1
    i = 0
with open('handshake/server_x25519.txt', 'r') as f1:
    for line in f1:
        x25519[i] = 10000000000 / int(line[42:49])
        i = i + 1
    i = 0
with open('handshake/server_swarm_ccold.txt', 'r') as f1:
    for line in f1:
        eng25519[i] = 10000000000 / int(line[42:49])
        i = i + 1
    i = 0
with open('handshake/server_all_openssl.txt', 'r') as f1:
    for line in f1:
        all_openssl[i] = 10000000000 / int(line[42:49])
        i = i + 1
    i = 0

with open('dot/unbound_p256.txt', 'r') as f1:
    for line in f1:
        value = line[42:50]
        if line[49] == ',':
            value = line[42:49]
        u_p256[i] = 10000000000 / int(value)
        i = i + 1
    i = 0
with open('dot/unbound_x25519.txt', 'r') as f1:
    for line in f1:
        value = line[42:50]
        if line[49] == ',':
            value = line[42:49]
        u_x25519[i] = 10000000000 / int(value)
        i = i + 1
    i = 0
with open('dot/unbound_swarm_ccold.txt', 'r') as f1:
    for line in f1:
        value = line[42:50]
        if line[49] == ',':
            value = line[42:49]
        u_eng25519[i] = 10000000000 / int(value)
        i = i + 1
    i = 0
with open('dot/unbound_all_openssl.txt', 'r') as f1:
    for line in f1:
        value = line[42:50]
        if line[49] == ',':
            value = line[42:49]
        u_all_openssl[i] = 10000000000 / int(value)
        i = i + 1
    i = 0

ecdf_p256 = sm.distributions.ECDF(p256)
ecdf_x25519 = sm.distributions.ECDF(x25519)
ecdf_eng25519 = sm.distributions.ECDF(eng25519)
ecdf_all_openssl = sm.distributions.ECDF(all_openssl)

u_ecdf_p256 = sm.distributions.ECDF(u_p256)
u_ecdf_x25519 = sm.distributions.ECDF(u_x25519)
u_ecdf_eng25519 = sm.distributions.ECDF(u_eng25519)
u_ecdf_all_openssl = sm.distributions.ECDF(u_all_openssl)

x_p256 = np.linspace(min(p256), max(p256))
x_x25519 = np.linspace(min(x25519), max(x25519))
x_eng25519 = np.linspace(min(eng25519), max(eng25519))
x_all_openssl = np.linspace(min(all_openssl), max(all_openssl))
print('p256 average: ', np.mean(p256))
print('x25519 average: ', np.mean(x25519))
print('eng25519 average: ', np.mean(eng25519))
print('all_openssl average: ', np.mean(all_openssl))

u_x_p256 = np.linspace(min(u_p256), max(u_p256))
u_x_x25519 = np.linspace(min(u_x25519), max(u_x25519))
u_x_eng25519 = np.linspace(min(u_eng25519), max(u_eng25519))
u_x_all_openssl = np.linspace(min(u_all_openssl), max(u_all_openssl))
print('u_p256 average: ', np.mean(u_p256))
print('u_x25519 average: ', np.mean(u_x25519))
print('eng25519 average: ', np.mean(u_eng25519))
print('u_all_openssl average: ', np.mean(u_all_openssl))

y_p256 = ecdf_p256(x_p256)
y_x25519 = ecdf_x25519(x_x25519)
y_eng25519 = ecdf_eng25519(x_eng25519)
y_all_openssl = ecdf_all_openssl(x_all_openssl)

u_y_p256 = u_ecdf_p256(u_x_p256)
u_y_x25519 = u_ecdf_x25519(u_x_x25519)
u_y_eng25519 = u_ecdf_eng25519(u_x_eng25519)
u_y_all_openssl = u_ecdf_all_openssl(u_x_all_openssl)

plt.figure(1, figsize=(13, 11), dpi=1000)
plt.xticks(np.arange(1100, 1750, 50), fontsize=25)
plt.yticks(np.arange(0, 1.1, 0.1), fontsize=25)
plt.xlim([1100, 1730])
plt.ylim([0, 1])
plt.grid(True)
plt.xlabel('#connections / second', fontsize=25)
plt.ylabel('Cumulative distribution: P(X ≤ x)', fontsize=25)

plt.plot(x_p256, y_p256, linewidth='3.2', label='P256', linestyle='-')
plt.plot(x_x25519, y_x25519, linewidth='3.2', label='X25519', linestyle='--')
plt.plot(x_eng25519, y_eng25519, linewidth='3.2', label='ENG25519', linestyle='-.')
plt.plot(x_all_openssl, y_all_openssl, linewidth='3.2', label='ALL-OpenSSL', linestyle=':')
plt.legend(loc='upper left', fontsize=25)
plt.savefig("handshake.jpg")
plt.savefig("handshake.pdf")
plt.show()

plt.figure(2, figsize=(13, 11), dpi=1000)
plt.xticks(np.arange(825, 1050, 25), fontsize=25)
plt.yticks(np.arange(0, 1.1, 0.1), fontsize=25)
plt.xlim([825, 1050])
plt.ylim([0, 1])
plt.grid(True)
plt.xlabel('#connections / second', fontsize=25)
plt.ylabel('Cumulative distribution: P(X ≤ x)', fontsize=25)

plt.plot(u_x_p256, u_y_p256, linewidth='3.2', label='P256', linestyle='-')
plt.plot(u_x_x25519, u_y_x25519, linewidth='3.2', label='X25519', linestyle='--')
plt.plot(u_x_eng25519, u_y_eng25519, linewidth='3.2', label='ENG25519', linestyle='-.')
plt.plot(u_x_all_openssl, u_y_all_openssl, linewidth='3.2', label='ALL-OpenSSL', linestyle=':')
plt.legend(loc='upper left', fontsize=25)
plt.savefig("dot.jpg")
plt.savefig("dot.pdf")
plt.show()
```
